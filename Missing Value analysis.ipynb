```python
data = pd.read_csv(r'C:\Users\ClinAsia1\Jupyter Notebooks\Datasets\Clinical dataset\Exchange csv\AE.csv')
data.head()
```

```python
data_columns= data.columns.tolist()
data_columns
```

# Display the count of missing values in dataset

```python
display_sum_missing_values = data.isna().sum()
missing_df = display_sum_missing_values.to_frame()
missing_df.columns = ['Count']
missing_df.index.names = ['Name']
missing_df['Name'] = data_columns
```

```python
missing_df.reset_index(drop=True, inplace=True)
```

```python
missing_df.sort_values(by=['Count'],ascending=False, inplace=True)
```

# Display the records indexes of missing values

```python
display = data.copy()
```

```python
display_miss_records_index = display[display.isna()].index.tolist()
display_miss_records_index
```

# Display the missing value records

```python
display_missing_records = display[display.isna().any(axis=1)]
display_missing_records
```

# Bar plot for number of missing values

```python
plot_missing= missing_df.plot.bar(x='Name', y='Count', rot=90, figsize=(10,6),use_index =True, sort_columns=True, legend=True)
```
